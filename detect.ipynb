```python
import gradio as gr
from fpdf import FPDF
import random

# Function to generate the report
def generate_report(level, reason):
    # Create PDF document
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    # Add a title
    pdf.set_font('Arial', 'B', 16)
    pdf.cell(200, 10, txt="Dark Circle Level Report", ln=True, align='C')

    # Add the level of dark circles and reasons
    pdf.ln(10)  # Line break
    pdf.set_font('Arial', '', 12)
    pdf.cell(200, 10, txt=f"Dark Circle Level: {level}", ln=True, align='L')
    pdf.cell(200, 10, txt=f"Reason: {reason}", ln=True, align='L')

    # Add the suggested steps to reduce dark circles
    steps = {
        'High': ['Get at least 7-8 hours of sleep', 'Use cold compress', 'Try dark circle creams'],
        'Moderate': ['Take frequent breaks from screens', 'Use a moisturizing cream', 'Try cucumber slices'],
        'Low': ['Get enough hydration', 'Maintain a balanced diet', 'Use sunscreen'],
        'No Dark Circle': ['Keep up the good habits!', 'Maintain hydration and rest']
    }
    pdf.ln(10)  # Line break
    pdf.cell(200, 10, txt="Suggested Steps to Reduce Dark Circles:", ln=True, align='L')
    for step in steps.get(level, []):
        pdf.cell(200, 10, txt=f"- {step}", ln=True, align='L')

    # Save PDF to a file
    file_path = '/content/dark_circle_report.pdf'
    pdf.output(file_path)

    return file_path

# Gradio Interface for Dark Circle Level detection and Report Generation
def detect_dark_circle(image):
    # Example dark circle levels based on the image
    levels = ['High', 'Moderate', 'Low', 'No Dark Circle']
    reasons = ['Lack of sleep', 'Stress and strain', 'Allergies', 'Good health']

    # Randomly assign a level and reason for demonstration purposes
    level = random.choice(levels)
    reason = reasons[levels.index(level)]

    # Generate the report PDF
    report_file = generate_report(level, reason)

    return f"Dark Circle Level: {level}", reason, report_file

# Create the Gradio Interface
iface = gr.Interface(
    fn=detect_dark_circle,
    inputs=gr.Image(type='pil'),
    outputs=[gr.Textbox(), gr.Textbox(), gr.File(label="Download Report")],
    live=True,
    title="Dark Circle Detection and Report Generation"
)

iface.launch()
```

```python
import cv2
from ultralytics import YOLO
import numpy as np

# Load the trained YOLO model
model = YOLO('runs/detect/train2/weights/best.pt')  # Make sure the path is correct

# Open webcam (0 is the default camera)
cap = cv2.VideoCapture(0)

# Check if the webcam is open
if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

while True:
    # Read frame from webcam
    ret, frame = cap.read()

    if not ret:
        print("Error: Failed to capture image.")
        break

    # Perform inference using YOLO
    results = model(frame)  # Inference on the current frame

    # Extract predictions from the results
    predictions = results[0].boxes.cls.tolist()

    # Display predictions (if any)
    if predictions:
        class_name = model.names[int(predictions[0])]
        print(f"Prediction: {class_name}")

        # Optionally, add advice (can be based on the class name like 'high', 'moderate', etc.)
        # You can integrate the advice dictionary logic here as we discussed earlier.

    # Display the results (image with predictions)
    cv2.imshow("Webcam Inference", results.render()[0])  # `results.render()` draws the boxes on the image

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
